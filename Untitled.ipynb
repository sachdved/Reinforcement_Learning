```python
relative_sizes = numpy.zeros(len(list_of_nums))
for i in range(len(list_of_nums)):
    relative_sizes[i] = list_of_nums[i]/list_of_nums[0]
```

```python
relative_sizes
```

```python
radii = numpy.sqrt(relative_sizes)
```

```python
radii[radii>10] = 10
```

```python
radii[radii>1.5] = 1.5
```
